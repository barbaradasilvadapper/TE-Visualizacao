## 2. Limpeza dos dados

```python
# Meses utilizados nas planilhas
meses = ["JAN", "FEV", "MAR", "ABR", "MAI", "JUN", "JUL", "AGO", "SET", "OUT", "NOV", "DEZ"]

mapa_meses = {
    "JAN": 1,
    "FEV": 2,
    "MAR": 3,
    "ABR": 4,
    "MAI": 5,
    "JUN": 6,
    "JUL": 7,
    "AGO": 8,
    "SET": 9,
    "OUT": 10,
    "NOV": 11,
    "DEZ": 12
}

categorias_lem = [
    "DESDOBRAMENTOS TÉCNICOS",
    "PROFISSIONALIZAÇÃO",
    "SAÚDE",
    "EDUCAÇÃO"
]
```

```python
def extrair_ano(nome_arquivo):
    match = re.search(r"20\d{2}", nome_arquivo)
    return int(match.group()) if match else None
```

```python
def limpar_texto(texto):
    if pd.isna(texto):
        return None
    
    texto = str(texto).strip()
    texto = re.sub(r"\s+", " ", texto)
    
    return texto
```

```python
def encontrar_linha_cabecalho(df_bruto):
    """
    Procura a linha que contém os meses JAN, FEV, MAR...
    """
    for indice, linha in df_bruto.iterrows():
        valores_linha = [limpar_texto(valor) for valor in linha.values]
        valores_linha = [valor for valor in valores_linha if valor is not None]

        qtd_meses = sum(valor in meses for valor in valores_linha)

        if qtd_meses >= 3:
            return indice

    return None
```

```python
def adicionar_categoria(df):
    categoria_atual = None
    categorias = []

    for indicador in df["indicador"]:
        indicador_limpo = limpar_texto(indicador)

        if indicador_limpo is not None and indicador_limpo.upper() in categorias_lem:
            categoria_atual = indicador_limpo.upper()
            categorias.append(categoria_atual)
        else:
            categorias.append(categoria_atual)

    df["categoria"] = categorias

    # Remove linhas que são somente títulos de categoria
    df = df[
        ~df["indicador"]
        .astype(str)
        .str.upper()
        .isin(categorias_lem)
    ]

    return df
```

```python
def ler_lem_mensal(arquivo):
    ano = extrair_ano(arquivo.name)

    # Lê a planilha sem cabeçalho para encontrar onde estão os meses
    df_bruto = pd.read_excel(
        arquivo,
        sheet_name="Planilha1",
        header=None
    )

    linha_cabecalho = encontrar_linha_cabecalho(df_bruto)

    if linha_cabecalho is None:
        raise ValueError("Não foi possível encontrar a linha de cabeçalho com os meses.")

    # Define a linha encontrada como cabeçalho
    novo_cabecalho = df_bruto.iloc[linha_cabecalho].apply(limpar_texto).tolist()

    # Dados começam na linha seguinte ao cabeçalho
    df = df_bruto.iloc[linha_cabecalho + 1:].copy()
    df.columns = novo_cabecalho

    # Renomeia a primeira coluna para indicador
    df = df.rename(columns={df.columns[0]: "indicador"})

    # Limpa nomes das colunas
    df.columns = [limpar_texto(col) for col in df.columns]

    # Mantém apenas indicador + meses
    colunas_meses_existentes = [mes for mes in meses if mes in df.columns]

    if len(colunas_meses_existentes) == 0:
        raise ValueError("Nenhuma coluna de mês foi encontrada.")

    df = df[["indicador"] + colunas_meses_existentes]

    # Remove linhas vazias
    df = df.dropna(how="all")
    df = df.dropna(subset=["indicador"])

    # Limpa os nomes dos indicadores
    df["indicador"] = df["indicador"].apply(limpar_texto)

    # Adiciona categoria
    df = adicionar_categoria(df)

    # Transforma de formato largo para longo
    df_longo = df.melt(
        id_vars=["categoria", "indicador"],
        value_vars=colunas_meses_existentes,
        var_name="mes",
        value_name="valor"
    )

    # Adiciona informações auxiliares
    df_longo["ano"] = ano
    df_longo["mes_numero"] = df_longo["mes"].map(mapa_meses)
    df_longo["fonte_arquivo"] = arquivo.name

    # Remove valores vazios
    df_longo = df_longo.dropna(subset=["valor"])

    # Converte para número
    df_longo["valor"] = pd.to_numeric(df_longo["valor"], errors="coerce")

    # Remove valores não numéricos, como X
    df_longo = df_longo.dropna(subset=["valor"])

    # Organiza colunas
    df_longo = df_longo[
        [
            "ano",
            "mes",
            "mes_numero",
            "categoria",
            "indicador",
            "valor",
            "fonte_arquivo"
        ]
    ]

    return df_longo
```

```python
arquivos_lem_mensal = [
    arquivo for arquivo in arquivos_excel
    if "anual" not in arquivo.name.lower()
]

bases_lem = []

for arquivo in arquivos_lem_mensal:
    try:
        df_tratado = ler_lem_mensal(arquivo)
        bases_lem.append(df_tratado)
        print(f"Arquivo processado com sucesso: {arquivo.name} | {df_tratado.shape[0]} linhas")
    except Exception as erro:
        print(f"Erro ao processar {arquivo.name}: {erro}")
```

```python
lem_consolidado = pd.concat(bases_lem, ignore_index=True)

lem_consolidado.head(20)
```
