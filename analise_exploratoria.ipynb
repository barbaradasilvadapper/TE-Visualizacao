## Análise exploratória do dataset LEM consolidado — 2021 a 2025

Este notebook realiza uma análise exploratória **somente sobre o dataset já padronizado** `lem_consolidado_2021_2025.csv`.

O objetivo desta etapa é verificar a estrutura, a qualidade e as principais características do conjunto de dados antes da definição final das visualizações interativas da solução.

## 1. Importação das bibliotecas

```python
import pandas as pd
import numpy as np
from pathlib import Path

import matplotlib.pyplot as plt
import seaborn as sns

from data_profiling import ProfileReport

sns.set_theme(style="whitegrid")
plt.rcParams["figure.figsize"] = (12, 6)
```

## 2. Leitura do dataset padronizado

O código abaixo procura o arquivo `lem_consolidado_2021_2025.csv` dentro de `dados_tratados/`;

```python
possiveis_caminhos = [
    Path("dados_tratados/lem_consolidado_2021_2025.csv"),
    Path("../dados_tratados/lem_consolidado_2021_2025.csv"),
]

caminho_dataset = None

for caminho in possiveis_caminhos:
    if caminho.exists():
        caminho_dataset = caminho
        break

if caminho_dataset is None:
    raise FileNotFoundError("Arquivo lem_consolidado_2021_2025.csv não encontrado. Verifique o caminho do dataset.")

print(f"Dataset encontrado em: {caminho_dataset}")

df = pd.read_csv(caminho_dataset)
df.head()
```

## 3. Ajuste de tipos e visão inicial

A coluna `data` é convertida para formato de data para permitir verificações temporais. As demais colunas são mantidas conforme o processamento anterior.

```python
df["data"] = pd.to_datetime(df["data"], errors="coerce")

df.info()
```

O `ydata-profiling` gera uma visão automática sobre o dataset, incluindo tipos de variáveis, valores ausentes, distribuições, frequências de categorias, estatísticas descritivas e alertas de qualidade.

## 5. Resumo estrutural do dataset

```python
resumo_estrutura = pd.DataFrame({
    "métrica": [
        "Quantidade de linhas",
        "Quantidade de colunas",
        "Anos distintos",
        "Meses distintos",
        "Categorias distintas",
        "Subcategorias distintas",
        "Indicadores distintos",
        "Indicadores completos distintos"
    ],
    "valor": [
        df.shape[0],
        df.shape[1],
        df["ano"].nunique(),
        df["mes"].nunique(),
        df["categoria"].nunique(),
        df["subcategoria"].nunique(dropna=True),
        df["indicador"].nunique(),
        df["indicador_completo"].nunique()
    ]
})

resumo_estrutura
```

## 6. Qualidade dos dados: valores ausentes

```python
nulos = (
    df.isna().sum()
    .reset_index()
    .rename(columns={"index": "coluna", 0: "qtd_nulos"})
)

nulos["percentual_nulos"] = (nulos["qtd_nulos"] / len(df) * 100).round(2)
nulos.sort_values("percentual_nulos", ascending=False)
```

```python
nulos_plot = nulos.sort_values("percentual_nulos", ascending=True)

plt.figure(figsize=(10, 5))
plt.barh(nulos_plot["coluna"], nulos_plot["percentual_nulos"])
plt.xlabel("Percentual de valores ausentes (%)")
plt.ylabel("Coluna")
plt.title("Percentual de valores ausentes por coluna")
plt.show()
```

## 7. Qualidade dos dados: duplicatas

```python
qtd_duplicatas = df.duplicated().sum()
percentual_duplicatas = round(qtd_duplicatas / len(df) * 100, 2)

pd.DataFrame({
    "métrica": ["Linhas duplicadas", "Percentual de duplicatas"],
    "valor": [qtd_duplicatas, percentual_duplicatas]
})
```

## 8. Cobertura temporal do dataset

Esta seção verifica se o dataset possui registros distribuídos entre os anos e meses esperados.

```python
cobertura_ano = (
    df.groupby("ano")
    .agg(
        qtd_registros=("valor", "count"),
        qtd_meses=("mes", "nunique"),
        qtd_indicadores=("indicador_completo", "nunique"),
        total_valor=("valor", "sum")
    )
    .reset_index()
)

cobertura_ano
```

```python
plt.figure(figsize=(8, 5))
plt.bar(cobertura_ano["ano"].astype(str), cobertura_ano["qtd_registros"])
plt.xlabel("Ano")
plt.ylabel("Quantidade de registros")
plt.title("Quantidade de registros por ano")
plt.show()
```

```python
cobertura_mes = (
    df.groupby(["ano", "mes_numero", "mes"])
    .size()
    .reset_index(name="qtd_registros")
    .sort_values(["ano", "mes_numero"])
)

cobertura_mes.head(20)
```

## 9. Composição por categoria e subcategoria

Esta análise mostra a composição estrutural da base, ou seja, quantos registros existem em cada área de indicadores.

```python
categoria_resumo = (
    df.groupby("categoria")
    .agg(
        qtd_registros=("valor", "count"),
        qtd_indicadores=("indicador_completo", "nunique"),
        total_valor=("valor", "sum")
    )
    .reset_index()
    .sort_values("qtd_registros", ascending=False)
)

categoria_resumo
```

```python
plt.figure(figsize=(10, 5))
plt.barh(categoria_resumo["categoria"], categoria_resumo["qtd_registros"])
plt.xlabel("Quantidade de registros")
plt.ylabel("Categoria")
plt.title("Quantidade de registros por categoria")
plt.gca().invert_yaxis()
plt.show()
```

```python
subcategoria_resumo = (
    df.dropna(subset=["subcategoria"])
    .groupby(["categoria", "subcategoria"])
    .agg(
        qtd_registros=("valor", "count"),
        qtd_indicadores=("indicador_completo", "nunique"),
        total_valor=("valor", "sum")
    )
    .reset_index()
    .sort_values(["categoria", "subcategoria"])
)

subcategoria_resumo
```

## 11. Valores zerados

Como os indicadores são mensais, valores iguais a zero podem representar ausência de ocorrência no mês. Esta verificação ajuda a entender a proporção de zeros no dataset.

```python
qtd_zeros = (df["valor"] == 0).sum()
percentual_zeros = round(qtd_zeros / len(df) * 100, 2)

pd.DataFrame({
    "métrica": ["Quantidade de valores zero", "Percentual de valores zero"],
    "valor": [qtd_zeros, percentual_zeros]
})
```

```python
zeros_por_categoria = (
    df.assign(valor_zero=df["valor"] == 0)
    .groupby("categoria")
    .agg(
        qtd_registros=("valor", "count"),
        qtd_zeros=("valor_zero", "sum")
    )
    .reset_index()
)

zeros_por_categoria["percentual_zeros"] = (
    zeros_por_categoria["qtd_zeros"] / zeros_por_categoria["qtd_registros"] * 100
).round(2)

zeros_por_categoria.sort_values("percentual_zeros", ascending=False)
```

```python
zeros_plot = zeros_por_categoria.sort_values("percentual_zeros", ascending=True)

plt.figure(figsize=(10, 5))
plt.barh(zeros_plot["categoria"], zeros_plot["percentual_zeros"])
plt.xlabel("Percentual de valores zero (%)")
plt.ylabel("Categoria")
plt.title("Percentual de valores zero por categoria")
plt.show()
```

## 12. Cardinalidade das variáveis categóricas

A cardinalidade mostra quantos valores distintos existem em cada coluna categórica. Isso ajuda a entender quais campos funcionam bem como filtros no futuro dashboard.

```python
colunas_categoricas = ["mes", "categoria", "subcategoria", "indicador", "indicador_completo"]

cardinalidade = pd.DataFrame({
    "coluna": colunas_categoricas,
    "valores_distintos": [df[col].nunique(dropna=True) for col in colunas_categoricas]
})

cardinalidade
```

```python
plt.figure(figsize=(10, 5))
plt.bar(cardinalidade["coluna"], cardinalidade["valores_distintos"])
plt.xlabel("Coluna")
plt.ylabel("Quantidade de valores distintos")
plt.title("Cardinalidade das variáveis categóricas")
plt.xticks(rotation=45, ha="right")
plt.show()
```
